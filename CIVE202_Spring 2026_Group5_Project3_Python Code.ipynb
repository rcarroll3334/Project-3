```python
def idm_acceleration(v, s, delta_v, v0, s0, T, a, b, delta):
    s_star = s0 + v*T + (v*delta_v) / (2*np.sqrt(a*b))
    s_star = max(s_star, s0)
    acceleration = a*(1 - ((v/v0)**delta) - ((s_star/s)**2))
    return acceleration
```

```python
v0 = 30    
s0 = 2     
T = 1.5    
a = 1.0    
b = 1.5    
delta = 4
test_v = 20
test_s = 25        
test_delta_v = 2
test_acc = idm_acceleration(test_v, test_s, test_delta_v, v0, s0, T, a, b, delta)
print(test_acc)
```

```python
trajectory_number = 1
trajectory_data = ngsim_data[ngsim_data["trajectory_number"] == trajectory_number].copy()
leader_position = trajectory_data["leader_position(m)"].values
leader_speed = trajectory_data["leader_speed(m/s)"].values
time = trajectory_data["Time"].values
dt = time[1] - time[0]
print("dt=", dt)
print("Number of time steps =", len(time))
```

```python
v0 = 30
s0 = 2
T = 1.5
a_param = 1.0 
b = 1.5
delta = 4
n_steps = len(trajectory_data)
simulated_follower_position = np.zeros(n_steps)
simulated_follower_speed = np.zeros(n_steps)
simulated_follower_acceleration = np.zeros(n_steps)
simulated_follower_position[0] = trajectory_data["follower_position(m)"].iloc[0]
simulated_follower_speed[0] = trajectory_data["follower_speed(m/s)"].iloc[0]
for i in range(n_steps - 1):
    gap = leader_position[i] - simulated_follower_position[i]
    delta_v = simulated_follower_speed[i] - leader_speed[i]
    gap = max(gap, 0.1)
    test_acc = idm_acceleration(simulated_follower_speed[i], gap, delta_v, v0, s0, T, a_param, b, delta)
    simulated_follower_acceleration[i] = test_acc
    simulated_follower_speed[i + 1] = simulated_follower_speed[i] + test_acc * dt
    simulated_follower_position[i + 1] = simulated_follower_position[i] + simulated_follower_speed[i] * dt
    simulated_follower_acceleration[-1] = simulated_follower_acceleration[-2]

print(simulated_follower_position[:5])
print(simulated_follower_speed[:5])
print(simulated_follower_acceleration[:5])
print("Simulation Complete!")
```
